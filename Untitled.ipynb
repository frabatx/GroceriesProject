```python
my_url = 'https://en.wikipedia.org/wiki/List_of_culinary_herbs_and_spices'
UClient =  uReq(my_url)
page_html = UClient.read()

# Extract code
page_soup = soup(page_html,"lxml")
# Extract tables
tables = page_soup.find_all("li")
```
